```python
from keras.models import load_model
model = load_model('Controls_and_Patients_small.h5')
```

```python
from keras.preprocessing.image import ImageDataGenerator
test_dir= '/Users/s436255/Downloads/Vgg16_Visualisation/Small Sample/Relocated_Data2/Test/'
test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = test_datagen.flow_from_directory(
         test_dir,            #Target directory
         target_size =(90, 80), 
         batch_size = 100,             
         class_mode='binary',  #Because you use binary_crossentropy loss, you need binary labels.
         shuffle=False)  # keep data in same order as labels


filenames = test_generator.filenames
nb_samples = len(filenames)

for data_batch, labels_batch in test_generator:
    print('data batch shape:', data_batch.shape)
    print('labels batch shape:', labels_batch.shape)
    break
```

```python
predict = model.predict_generator(test_generator,steps = nb_samples/100)
print("Accuracy = ",predict[1])
```
